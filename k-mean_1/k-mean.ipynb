```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings( "ignore" )
pd.options.display.max_columns = None
import matplotlib.pyplot as plt
import matplotlib as mpl
%pylab inline
plt.style.use( 'seaborn-dark' )
plt.rc( 'font', size = 14)
```

```python
from sklearn.datasets import make_blobs
seed = 10
n_data = 200
n_clusters = 4
n_centers = 4

blobs, blob_labels = make_blobs( n_samples=n_data, n_features=2,
                                 centers=n_centers, random_state=seed,
                                 cluster_std=2)
blobs *= 0.75
blobs += 1.1

plt.figure( figsize=(4, 4))
plt.scatter( blobs[:, 0], blobs[:, 1], 
             s=70, c="#000F99", edgecolors="white", linewidths=0.8)
plt.xlabel("feature 1")
plt.ylabel("feature 2")

plt.grid( lw = 2)
plt.xlim([-10, 10])
plt.ylim([-10, 10])
```

```python
from sklearn.cluster import KMeans

clusters = KMeans( n_clusters = 6, random_state = 0, init='random').fit_predict( blobs )

plt.figure( figsize=(4, 4))

plt.scatter( blobs[:, 0], blobs[:, 1],
             c=clusters, s=50, cmap='tab20',
             edgecolors="black", linewidths=1.0)

plt.xlabel("feature 1")
plt.ylabel("feature 2")

plt.grid(lw=2)
plt.xlim([-10, 10])
plt.ylim([-10, 10])
```
